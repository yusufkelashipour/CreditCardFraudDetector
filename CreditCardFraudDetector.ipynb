```python
#Import Dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Load dataset to pandas dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
#Display first and last 5 rows
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
credit_card_data.info()
```

```python
#Check for number of missing values in each column
credit_card_data.isnull().sum()
```

```python
#Check distribution of legit vs fraud transactions | 0 is legit, 1 is fraud
credit_card_data['Class'].value_counts()
```

```python
#Seperate the transactions for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#Statistical measures of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#Compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

```python
#Create sample dataset containing similar distribution of legit and fraud transactions
legit_sample = legit.sample(n=492)
```

```python
#Concatenate sample dataset to fraud dataset
new_dataset = pd.concat([legit_sample,fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#Split data into features & targets
X = new_dataset.drop(columns='Class', axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Split data into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size= 0.20, stratify = Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training | Logistic Regression

```python
model = LogisticRegression(max_iter=10000)
```

```python
#Training the logistic regression model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

```python
#Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training Data: ', training_data_accuracy)
```

```python
#Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test Data: ', test_data_accuracy)
```
